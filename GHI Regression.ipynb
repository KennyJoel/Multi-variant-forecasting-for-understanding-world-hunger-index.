Importing the necessary libraries: Pandas for data handling and Linear Regression model from Scikit-Learn for machine learning.

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

Loading the data from an Excel file and printing the first few rows of the data using the head() function to inspect the data.

```python
df = pd.read_excel("F:/SEM 6/ML/2nd Evalution/GHIData.xlsx")
df.head()
```

Importing the Linear Regression model and initializing an instance of the model.

```python
from sklearn.linear_model import LinearRegression
lin_model=LinearRegression()
```

Importing the Random Forest Regressor model and initializing an instance of the model with 100 trees and max_features set to 3.

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(n_estimators=100,max_features=3, random_state=1)
```

Extracting the features and target variable from the loaded data, converting them to NumPy arrays and reshaping them into arrays with one column. Then concatenating the feature arrays into a final feature matrix using np.concatenate() function.

```python
import numpy as np
x1,x2,x3,y=df['A'],df['B'],df['C'],df['D']
x1,x2,x3,y=np.array(x1),np.array(x2),np.array(x3),np.array(y)
x1,x2,x3,y=x1.reshape(-1,1),x2.reshape(-1,1),x3.reshape(-1,1),y.reshape(-1,1)
final_x=np.concatenate((x1,x2,x3),axis=1)
print(final_x)
```

Splitting the final feature matrix and target variable into training and testing sets with the last 30 observations as the test set.

```python
X_train,X_test,y_train,y_test=final_x[:-30],final_x[-30:],y[:-30],y[-30:]
```

Fitting the Random Forest Regressor model and Linear Regression model on the training data.

```python
model.fit(X_train,y_train)
lin_model.fit(X_train,y_train)
```

Making predictions using the Random Forest Regressor model on the test data, plotting the predicted values against the actual values using Matplotlib.

```python
pred=model.predict(X_test)
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (12,8)
plt.plot(pred,label='Random_Forest_Predictions')
plt.plot(y_test,label='Actual ')
plt.legend(loc="upper left")
plt.show()
```

Making predictions using the Linear Regression model on the test data, plotting the predicted values against the actual values using Matplotlib.

```python
lin_pred=lin_model.predict(X_test)
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (11,6)
plt.plot(lin_pred,label='Linear_Regression_Predictions')
plt.plot(y_test,label='Actual')
plt.legend(loc="upper left")
plt.show()
     
```

Calculating the Root Mean Squared Error (RMSE) between the predicted and actual values for both the models using the mean_squared_error() function from Scikit-Learn and printing the RMSE values for both the models.

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
rmse_rf=sqrt(mean_squared_error(pred,y_test))
rmse_lr=sqrt(mean_squared_error(lin_pred,y_test))
  
```

```python
print('Mean Squared Error for Random Forest Model is:',rmse_rf)
print('Mean Squared Error for Linear Regression Model is:',rmse_lr)
     
```
