## Filtering Data

```python
#importing regular expression function and trying to play the words which has "in"
import re
df.loc[df['TITLE'].str.contains('In[a-z]*', flags=re.I, regex=True)]
```

```python
#trying to play around which words of titles starts with 'in'
df.loc[df['TITLE'].str.contains('^In[a-z]*', flags=re.I, regex=True)]
```

## Aggregate Statistics (Groupby)


```python
df = pd.read_csv('F:/SEM 6/ML LAB/ML LAB 1/archive/Best Movies Netflix.csv')
df
```

```python
df['count'] = 1
df
```

```python
df.groupby(['TITLE']).count()
df.groupby(['MAIN_GENRE', 'SCORE']).count()['count']  #get coutn based on multiple attributes in dataset
```
